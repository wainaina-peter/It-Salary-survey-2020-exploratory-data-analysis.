```python
#all relevant imports
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#loading data
mydata = pd.read_csv("IT Salary Survey EU  2020.csv")
```

```python
#to find missing values
mydata.isna()
```

```python
#to find missing values per column, and include the sum() function
mydata.isna().sum()
```

```python
#checking count for genders
mydata["Gender"].value_counts()
```

```python
#visualizing the gender count
sns.countplot(data=mydata,x="Gender")
plt.title("IT Professions per Gender")
plt.ylabel("Number of People")
plt.show()
```

```python
#Exploring worker distribution by their years of experience
#Define plot size
plt.figure(figsize=[6,6])

#Define column to use
data = mydata["Total years of experience"].value_counts(normalize=True)

#Define labels
labels = [f"Label {i}" for i in range(len(data))]

#Define color palette
colors = sns.color_palette('pastel')

#Create pie chart
plt.pie(data,labels=labels,colors=colors, autopct='%.0f%%')
plt.title("Distribution of workers by their years of experience")
plt.show()
```

```python
#Exploring worker distribution by their seniority level
#Define plot size
plt.figure(figsize=[6,6])

#Define column to use
data = mydata["Seniority level"].value_counts(normalize=True)

#Define labels
labels = [f"Label {i}" for i in range(len(data))]

#Define color palette
colors = sns.color_palette('pastel')

#Create pie chart
plt.pie(data,labels=labels,colors=colors, autopct='%.0f%%')
plt.title("Distribution of workers by their seniority level")
plt.show()
```
